```python
import os
import nbformat

def clear_outputs_in_notebooks(filepath):
    # Charger le notebook
    with open(filepath, "r", encoding="utf-8") as f:
        nb = nbformat.read(f, as_version=4)
    
    # Parcourir chaque cellule et supprimer les sorties
    for cell in nb.cells:
        if "outputs" in cell:
            cell["outputs"] = []
    
    # Sauvegarder le notebook sans les sorties
    with open(filepath, "w", encoding="utf-8") as f:
        nbformat.write(nb, f)

def parcourir_repertoire(repertoire):
    # Parcourir tous les fichiers et répertoires dans le répertoire donné
    for nom in os.listdir(repertoire):
        # Construire le chemin complet du fichier ou du répertoire
        chemin = os.path.join(repertoire, nom)
        # Vérifier si c'est un répertoire
        if os.path.isdir(chemin):
            print(f"Répertoire trouvé : {chemin}")
            # Si c'est un répertoire, appeler récursivement la fonction sur ce répertoire
            parcourir_repertoire(chemin)
        elif chemin.endswith(".ipynb"):
            print(f"Fichier trouvé : {chemin}")
            # Si c'est un fichier notebook, supprimer les sorties
            clear_outputs_in_notebooks(chemin)

# Exemple d'utilisation
directory_path = "/gpfs/commons/groups/gursoy_lab/mstoll/codes"
parcourir_repertoire(directory_path)
```
